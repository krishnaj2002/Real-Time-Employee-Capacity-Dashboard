```python
from fpdf import FPDF
import pandas as pd
import requests
from io import StringIO
from datetime import datetime, timedelta
import os
import webbrowser

class PDF(FPDF):
    def header(self):
        self.image("logo.png", 10, 8, 30)
        self.set_font("Arial", 'B', 16)
        self.set_text_color(0, 51, 102)
        self.cell(0, 10, "Tekhnologia Innovations Private Limited", ln=True, align="C")
        self.set_font("Arial", 'B', 12)
        self.cell(0, 10, "Weekly Employee Capacity Report", ln=True, align="C")
        self.ln(5)

    def footer(self):
        self.set_y(-10)
        self.set_font("Arial", 'I', 8)
        self.set_text_color(169, 169, 169)
        self.cell(0, 10, "Tekhnologia Innovations Pvt. Ltd. | +91 77758 18701 | info@tekhnologiaindia.com", 0, 0, 'C')
        self.ln(4)
        self.cell(0, 10, f'Page {self.page_no()}', align="C")

def generate_report():
    try:
        print("üì• Fetching Google Sheets data...")

        # Google Sheets to CSV
        sheet_url = "https://docs.google.com/spreadsheets/d/1y-aWZedtwLtd6f1pQo3DpJMyYO0PO7tgkfWZ-UGzSxI/edit?usp=sharing"
        csv_url = sheet_url.replace("/edit?usp=sharing", "/export?format=csv")
        response = requests.get(csv_url)
        response.raise_for_status()
        df = pd.read_csv(StringIO(response.text))
        df.columns = df.columns.str.strip()
        df.fillna("-", inplace=True)

        # Filter required columns
        report_df = df[["Emp Code", "Emp Name", "Designation", "Projects", "Allocation", "Priority", "Deadline"]].copy()
        report_df["Clean Allocation"] = report_df["Allocation"].replace("-", "0").str.replace("%", "").astype(float)

        # Date range
        today = datetime.today()
        last_sunday = today - timedelta(days=today.weekday() + 1)
        this_saturday = last_sunday + timedelta(days=6)
        report_start = last_sunday.strftime("%d/%m/%Y")
        report_end = this_saturday.strftime("%d/%m/%Y")
        pdf_filename = f"Employee_Capacity_Report_{report_end.replace('/', '-')}.pdf"

        print("üìù Creating landscape PDF...")
        pdf = PDF(orientation='L', unit='mm', format='A4')
        pdf.add_page()

        # Date range text
        pdf.set_font("Arial", 'I', 11)
        pdf.set_text_color(100, 100, 100)
        pdf.cell(0, 10, f"Report from {report_start} to {report_end}", ln=True, align="C")
        pdf.ln(3)

        # Summary Stats
        total_employees = len(report_df)
        fully_allocated = (report_df["Clean Allocation"] == 100).sum()
        partially_allocated = (report_df["Clean Allocation"] < 100).sum()
        high_priority = (report_df["Priority"].str.lower().str.strip() == "high").sum()
        medium_priority = (report_df["Priority"].str.lower().str.strip() == "medium").sum()
        low_priority = (report_df["Priority"].str.lower().str.strip() == "low").sum()

        pdf.set_font("Arial", '', 10)
        pdf.set_text_color(0)
        pdf.multi_cell(0, 8,
            f"Total Employees: {total_employees}    |    Fully Allocated: {fully_allocated}    |    Partially Allocated: {partially_allocated}\n"
            f"High Priority: {high_priority}    |    Medium Priority: {medium_priority}    |    Low Priority: {low_priority}",
            align='L'
        )
        pdf.ln(4)

        # Table Headers
        headers = ["Emp Code", "Emp Name", "Designation", "Projects", "Allocation", "Priority", "Deadline"]
        col_widths = [28, 45, 35, 100, 25, 25, 35]
        table_width = sum(col_widths)
        margin_x = (pdf.w - table_width) / 2
        pdf.set_x(margin_x)

        pdf.set_fill_color(0, 51, 102)
        pdf.set_text_color(255)
        pdf.set_font("Arial", 'B', 10)
        for i, header in enumerate(headers):
            pdf.cell(col_widths[i], 10, header, border=1, align='C', fill=True)
        pdf.ln()

        # Table Rows
        pdf.set_font("Arial", '', 9)
        pdf.set_text_color(0)
        for idx, row in report_df.iterrows():
            pdf.set_x(margin_x)
            fill = idx % 2 == 0
            pdf.set_fill_color(245, 245, 245) if fill else pdf.set_fill_color(255, 255, 255)
            for i, key in enumerate(headers):
                align = 'C' if key in ['Allocation', 'Priority', 'Deadline'] else 'L'
                pdf.cell(col_widths[i], 8, str(row[key]), border=1, align=align, fill=True)
            pdf.ln()

        # Save PDF
        pdf.output(pdf_filename)
        print(f"‚úÖ Report generated: {pdf_filename}")

        # Preview (OS-specific)
        path = os.path.abspath(pdf_filename)
        if os.name == 'nt':
            os.startfile(path)
        else:
            webbrowser.open_new(f"file://{path}")

    except Exception as e:
        print("‚ùå Error:", e)

# Run once
generate_report()
```
